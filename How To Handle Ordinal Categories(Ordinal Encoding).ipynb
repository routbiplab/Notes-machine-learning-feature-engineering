```python
import numpy as np
import pandas as pd
import datetime
```

```python
#here we are going to days of the weeke and we will label it based on their occurance in an week
```

```python
today = datetime.datetime.today()
today
```

```python
#here we took last 20 days datetime
days_list = [today-datetime.timedelta(x) for x in range(0,20)]
days_list
```

```python
df = pd.DataFrame(days_list)
df
```

```python
df.columns = ["Days"]
```

```python
df
```

```python
# extract days name from datetime
df["Day Name"] = df["Days"].dt.weekday
```

```python
df
```

```python
day_names = dict({0:"Monday",1:"Tuesday",2:"Wednesday",3:"Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
```

```python
day_names
```

```python
df["Day Name"]
```

```python
df
```

```python
df["Day Name"] = df["Day Name"].map(day_names)
```

```python
df
```

```python
# mapping ordinal value to day names
day_ordinal = {"Monday":1,"Tuesday":2,"Wednesday":3,"Thursday":4,"Friday":5,"Saturday":6,"Sunday":7}
day_ordinal
```

```python
df["Day Ordinal"] = df["Day Name"].map(day_ordinal)
```
