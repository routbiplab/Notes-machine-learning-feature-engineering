# Continous variables plot - joint plot, pair plot and dist plot

```python
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.corr()
```

# Heat Map

```python
sns.heatmap(data.corr())
```

# Joint Plot

## Bi-variate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="reg")
```

# Pair Plot
## Multi variate analysis

```python
sns.pairplot(data,hue="smoker")
```

# Dist plot

```python
sns.displot(data,kde=True,bins=2)
```

# Categorical variables - Box, violin, count, bar plots

## Count Plot

```python
df=data
sns.countplot("sex",data=df)
```

```python
sns.countplot("day",data=df)
```

```python
sns.countplot(y="smoker",data=df)
```

## Bar plot

```python
sns.barplot(x="sex",y="tip",data=df)
```

```python
sns.barplot(x="day",y="total_bill",data=df)
```

## Box Plot (x or y needs to be numeric)

```python
# x=categorical , y= numerical
sns.boxplot(x="day",y="total_bill",data=df,palette="rainbow")
```

```python
sns.boxplot(x="day",y="tip",data=df)
```

```python
sns.boxplot(data=df,orient="h")
```

```python
df.head()
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=df)
```

## Violin Plot

```python
sns.violinplot(x="sex",y="total_bill",data=df)
```
