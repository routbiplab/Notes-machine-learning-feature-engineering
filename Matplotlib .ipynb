```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c="g")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph in 2-D")
plt.savefig("2d-graph.png")
```

```python
y=x*x
```

```python
plt.plot(x,y,"r*--",markersize=12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("new 2-d graph")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"ro")
plt.subplot(2,2,2)
plt.plot(x,y,"go")
plt.subplot(2,2,3)
plt.plot(x,y,"yo")
plt.subplot(2,2,4)
plt.plot(x,y,"bo")
```

```python
x=np.arange(0,4*np.pi,.2)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine Graph")
```

```python
x=np.arange(0,4*np.pi,.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,"r--")
plt.title("Sine Curve")
plt.subplot(2,1,2)
plt.plot(x,y_cos,"g--")
plt.title("Cos Curve")
```

```python
x=np.array([23,4,5,4,6,7,6,7,8,23])
plt.hist(x,bins=8)
plt.title("histogram")
```

```python
data=[np.random.normal(1,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
gas=["nitrogen","oxygen","carbon-dioxide"]
colors=("blue","green","red")
data=[780,200,20]
explode=(.1,0,0)
plt.pie(data,labels=gas,colors=colors,explode=explode,shadow=True,autopct="%1.1f%%")
```
