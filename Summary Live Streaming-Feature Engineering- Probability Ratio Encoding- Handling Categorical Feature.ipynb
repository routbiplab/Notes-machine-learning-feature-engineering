```python
import pandas as pd
import numpy as np
```

# Probability Ratio Encoding- Handling Categorical Feature

```python
df=pd.read_csv("titanic_train.csv",usecols=["Cabin","Survived"])
df.head()
```

```python
df.Cabin.fillna("Missing",inplace=True)
df.head()
```

```python
# taking the 1st letter from cabin
df["Cabin"]=df["Cabin"].astype(str).str[0]
df.head(10)
```

```python
# group the cabin based on survival mean
prob_df=df.groupby("Cabin")["Survived"].mean()
prob_df=pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
# prob of died based on cabin values
prob_df["Died"]=1-prob_df["Survived"]
prob_df
```

```python
# probability ratio encoding
prob_df["Probability_Ratio"]=prob_df["Survived"]/prob_df["Died"]
prob_df
```

```python
prob_dict=prob_df["Probability_Ratio"].to_dict()
prob_dict
```

```python
df["Cabin_encoded"]=df["Cabin"].map(prob_dict)
df.head(20)
```
